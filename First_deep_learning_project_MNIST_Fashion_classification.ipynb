```python
'''
class_labels:
  0=>  T-shirt/top
  1 => Trouser
  2 => Pullover
  3 => Dress
  4 => Coat
  5 => Sandal
  6 => Shirt
  7 => Sneaker
  8 => Bag
  9 => Ankle boot
  Classify the given input from above class using Neural Network
  image shape 28*28 pixel (Gray scale)

'''
```

```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.fashion_mnist.load_data()
x_train.shape, y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
class_lables = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
'''
0 => T-shirt/top
1 => Trouser
2 => Pullover
3 => Dress
4 => Coat
5 => Sandal
6 => Shirt
7 => Sneaker
8 => Bag
9 => Ankle boot
'''
```

```python
plt.imshow(x_train[0], cmap = "Greys")
```

```python
plt.figure(figsize=(16,16))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i],cmap='Greys')
    plt.axis('off')
    plt.title(class_lables[y_train[i]]+"="+str(y_train[i]),fontsize=20)
    '''
    0=>T-shirt/top
    1=>Trouser
    2=>Pullover
    3=>Dress
    4=>Coat
    5=>Sandal
    6=>Shirt
    7=>Sneaker
    8=>Bag
    9=>Ankle boot
    '''
```

```python
x_train = x_train/255
x_test  = x_test/255
x_train[0]
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(units=32,activation='relu'),
    keras.layers.Dense(units=32,activation='softmax')
])
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=1)
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[0].round(2)
```
